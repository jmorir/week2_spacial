## Importación de data

Se procede a importar la geodata para ser comprendida en el análisis y psoteriormente seleccionada para la creación de los mapas relevantes para el caso uruguayo.

```python
import os, geopandas as gpd

countries=gpd.read_file(os.path.join("maps","World_Countries","World_Countries.shp"))
rivers=gpd.read_file(os.path.join("maps","World_Hydrography","World_Hydrography.shp"))
cities=gpd.read_file(os.path.join("maps","World_Cities","World_Cities.shp"))
```

```python
countries.head()
```

Se revisa que no hayan valores perdidos

```python
countries[countries.isna().any(axis=1)]
```

```python
countries.info()
```

Se procede a realizar la graficación de los mapas. A continuación, se realizará la graficación de los polígonos, puntos y líneas para posteriormente unificar la data. Sin embargo, antes se debe revisar si los mapas están en la misma proyección.

```python
countries.crs==cities.crs==cities.crs
```

```python
countries.plot(facecolor="lightblue",
               edgecolor='black',
               linewidth=0.1)
```

```python
rivers.plot(edgecolor='blue',
            linewidth=1,
            linestyle='dotted')
```

```python
cities.plot(marker='.',
            color='red',
            markersize=1,
            alpha=0.5)
```

```python
base = countries.plot(facecolor="white",
                      edgecolor='black',
                      linewidth=0.1,
                      figsize=(12,12))

rivers.plot(edgecolor='blue', linewidth=0.4,
            ax=base)

cities.plot(marker='.', color='red', markersize=1,
            ax=base)
```

```python
countries.to_file(os.path.join("data","worldMap.gpkg"),layer='countryBorders', driver="GPKG")
rivers.to_file(os.path.join("data","worldMap.gpkg"),layer='riverLines', driver="GPKG")
cities.to_file(os.path.join("data","worldMap.gpkg"),layer='cityPoints', driver="GPKG")
```

## Selección del país - Uruguay

A continuación se filtrarán los países para seleccionar únicamente Uruguay. De esta manera se utilizará la función filter y clip para trabajar únicamente con la información de Uruguay.

```python
uruguay=countries[countries.COUNTRY=='Uruguay']
```

```python
citiesUruguay_clipped = gpd.clip(gdf=cities,
                          mask=uruguay)
riversUruguay_clipped = gpd.clip(gdf=rivers,
                               mask=uruguay)
```

```python
base = uruguay.plot(facecolor="skyblue", edgecolor='black', linewidth=0.4,figsize=(5,5))
citiesUruguay_clipped.plot(marker='+', color='red', markersize=20,
                    ax=base)
riversUruguay_clipped.plot(edgecolor='blue', linewidth=1,
                    ax=base)
```

### Reproyección
El CRS de Uruguay según [ESPEG.IO](https://epsg.io/?q=uruguay+kind%3APROJCRS) es de 5382. Por ello se realizarán las verificaciones de acuerdo al dato del centroide.


```python
uruguay.centroid
```

```python
uruguay.to_crs(5382).crs.axis_info
```

```python
uruguay.to_crs(5382).centroid
```

Se replotea para demostrar el centroide en el caso de Uruguay

```python
base5382=uruguay.to_crs(5382).plot()
uruguay.to_crs(5382).centroid.plot(color='red',ax=base5382)
```

Se procede a guardar la información de la reproyección del mapa de Uruguay y los geopackages por capas en la carpeta data

```python
uruguay_5382=uruguay.to_crs(5382)
cities_uruguay_5382=citiesUruguay_clipped.to_crs(uruguay_5382.crs)
rivers_uruguay_5382=riversUruguay_clipped.to_crs(uruguay_5382.crs)
```

```python
import os
uruguay_5382.to_file(os.path.join("data","uruguayMaps_5382.gpkg"), layer='country', driver="GPKG")
cities_uruguay_5382.to_file(os.path.join("data","uruguayMaps_5382.gpkg"), layer='cities', driver="GPKG")
rivers_uruguay_5382.to_file(os.path.join("data","uruguayMaps_5382.gpkg"), layer='rivers', driver="GPKG")
```

```python
uruguay_5382.centroid
```

```python
uruguay_5382.centroid.to_file(os.path.join("data","uruguayMaps_5382.gpkg"), layer='centroid', driver="GPKG")
```
