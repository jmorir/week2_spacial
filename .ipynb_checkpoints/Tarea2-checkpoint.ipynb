## Importación de data

Se procede a importar la geodata para ser comprendida en el análisis y psoteriormente seleccionada para la creación de los mapas relevantes para el caso uruguayo.

```python
import os, geopandas as gpd

countries=gpd.read_file(os.path.join("data","World_Countries","World_Countries.shp"))
rivers=gpd.read_file(os.path.join("data","World_Hydrography","World_Hydrography.shp"))
cities=gpd.read_file(os.path.join("data","World_Cities","World_Cities.shp"))
```

```python
countries.crs==cities.crs==cities.crs
```

```python
countries.to_file(os.path.join("data","worldMap.gpkg"),layer='countryBorders', driver="GPKG")
rivers.to_file(os.path.join("data","worldMap.gpkg"),layer='riverLines', driver="GPKG")
cities.to_file(os.path.join("data","worldMap.gpkg"),layer='cityPoints', driver="GPKG")
```

## Selección del país - Uruguay

A continuación se filtrarán los países para seleccionar únicamente Uruguay. De esta manera se utilizará la función filter y clip para trabajar únicamente con la información de Uruguay.

```python
uruguay=countries[countries.COUNTRY=='Uruguay']
```

```python
citiesUruguay_clipped = gpd.clip(gdf=cities,
                          mask=uruguay)
riversUruguay_clipped = gpd.clip(gdf=rivers,
                               mask=uruguay)
```

```python
base = uruguay.plot(facecolor="skyblue", edgecolor='black', linewidth=0.4,figsize=(5,5))
citiesUruguay_clipped.plot(marker='+', color='red', markersize=20,
                    ax=base)
riversUruguay_clipped.plot(edgecolor='blue', linewidth=1,
                    ax=base)
```

### Reproyección
El CRS de Uruguay según [ESPEG.IO](https://epsg.io/?q=uruguay+kind%3APROJCRS) es de 5382. Por ello se realizarán las verificaciones de acuerdo al dato del centroide.


```python
uruguay.centroid
```

```python
uruguay.to_crs(5382).crs.axis_info
```

```python
uruguay.to_crs(5382).centroid
```

Se replotea para demostrar el centroide en el caso de Uruguay

```python
base5382=uruguay.to_crs(5382).plot()
uruguay.to_crs(5382).centroid.plot(color='red',ax=base5382)
```

Se procede a guardar la información de la reproyección del mapa de Uruguay y los geopackages por capas en la carpeta data

```python
uruguay_5382=uruguay.to_crs(5382)
cities_uruguay_5382=citiesUruguay_clipped.to_crs(uruguay_5382.crs)
rivers_uruguay_5382=riversUruguay_clipped.to_crs(uruguay_5382.crs)
```

```python
import os
uruguay_5382.to_file(os.path.join("data","uruguayMaps_5382.gpkg"), layer='country', driver="GPKG")
cities_uruguay_5382.to_file(os.path.join("data","uruguayMaps_5382.gpkg"), layer='cities', driver="GPKG")
rivers_uruguay_5382.to_file(os.path.join("data","uruguayMaps_5382.gpkg"), layer='rivers', driver="GPKG")
```

```python
uruguay_5382.centroid
```

```python
uruguay_5382.centroid.to_file(os.path.join("data","uruguayMaps_5382.gpkg"), layer='centroid', driver="GPKG")
```

## Adición de información espacial (aeropuertos) 

En esta sección se agregará la información sobre aeropuertos para complementar la graficación de los mapas

```python
import pandas as pd
infoairports=pd.read_csv(os.path.join("data","uy-airports.csv"))

# some rows

infoairports.iloc[[0,1,2,3,-4,-3,-2,-1],:] #head and tail
```

```python
infoairports.columns
```

```python
keep=['name','type','latitude_deg', 'longitude_deg','elevation_ft','region_name','municipality']
infoairports=infoairports.loc[:,keep]
```

```python
numericCols=['latitude_deg', 'longitude_deg','elevation_ft']
infoairports[numericCols]=infoairports.loc[:,numericCols].apply(lambda x:pd.to_numeric(x))

infoairports.info()
```

```python
base = uruguay_5382.plot(color='white', edgecolor='black')

infoairports.plot.scatter(x = 'longitude_deg', y = 'latitude_deg',ax=base)
```

```python
airports=gpd.GeoDataFrame(data=infoairports.copy(),
                 geometry=gpd.points_from_xy(infoairports.longitude_deg,
                                             infoairports.latitude_deg),
                 crs=uruguay.crs.to_epsg())
```

```python
base = uruguay_5382.plot(color='white', edgecolor='black')
airports.plot(ax=base)
```

```python
type(airports), type(infoairports)
```

```python
airports_5382=airports.to_crs(5382)

base = uruguay_5382.plot(color='white', edgecolor='black')
airports_5382.plot(ax=base, markersize=10, color='green')
```

```python
airports_5382['type'].value_counts()
```

```python
airports_5382.rename(columns={'type':'kind'},inplace=True)
```

```python
airports_5382.to_file(os.path.join("data","uruguayMaps_5641.gpkg"), layer='airports', driver="GPKG")
```

### Agregar a los mapas anteriores

A continuación, se agregará el mapa de los aeropuertos a los mapas ya realizados (ciudades y ríos)

```python
base = uruguay_5382.plot(facecolor="skyblue", edgecolor='black', alpha=0.5, linewidth=0.4,figsize=(5,5))
cities_uruguay_5382.plot(marker='+', color='red', markersize=50,
                    ax=base)
rivers_uruguay_5382.plot(edgecolor='blue', linewidth=1,
                    ax=base)
airports_5382.plot(ax=base, markersize=10, color='green')
```

## Agregar límites administrativos de Uruguay

El territorio adminsitrativo de Uruguay está políticamente dividio en 19 departamentos. Para el ejercicio, se ejecutará la división por departamentos ya que no existe un equivalente apropiado a provincias.

```python
limites = gpd.read_file(os.path.join("data", "limitesDepartamentales_SHP", "Departamentos.shp"))
```

```python
limites_5382 = limites.to_crs(5382)
```

## Creación del Mapa de Uruguay

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10, 10))

base = uruguay_5382.plot(ax=ax, facecolor="skyblue", edgecolor='black', alpha=0.5, linewidth=0.4)
cities_uruguay_5382.plot(ax=base, marker='+', color='red', markersize=50, label='Ciudades')
rivers_uruguay_5382.plot(ax=base, edgecolor='blue', linewidth=1, label='Ríos')
airports_5382.plot(ax=base, markersize=10, color='green', label='Aeropuertos')

limites_5382.plot(ax=base, edgecolor='black', linewidth=0.1, label='Límites')

ax.set_title("Mapa de Uruguay con Ciudades, Ríos, Aeropuertos y Límites", fontsize=16)

ax.legend(loc='upper right', fontsize=12)

plt.show()
```
